```python
# Import the necessary libraries
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from pandas import ExcelWriter
from pandas import ExcelFile
```

```python
# Import the Travel Market data
tm = pd.read_excel('APAC_Travel_Market.xlsx', sheet_name='APAC Travel Market')
```

```python
# Working with Data Frame
type(tm)
```

```python
# Looks good!
tm.head()
```

```python
tm = tm.sort_values(2000, ascending = False).reset_index(drop=True)
tm.head()
```

```python
plt.rcParams['figure.figsize'] = [8, 5]
fig, ax = plt.subplots() # Create one figure with one axis

ind = np.arange(len(tm['Region']))    # the x locations for the groups
width = 0.35        # the width of the bars
rects1 = ax.bar(ind - width/2, tm[2000], width, color='lightblue', bottom=0, tick_label=tm['Region'])
rects2 = ax.bar(ind + width/2, tm[2016], width, color='blue', bottom=0)
ax.get_yaxis().set_visible(False) # Remove thee y-axis
ax.xaxis.set_ticks_position('none') # Remove the x-axis ticks
ax.legend({2000,2016})
plt.box(False) # Remove the plot box
plt.suptitle('Asia Pacific (APAC) Travel Market' , fontsize = 18, ha='right', y = 1.05, x=0.58) # Title plot.  x,y = (0,0) = bottom left, x,y= (1,1) = top right 
plt.title('Big and getting bigger', y = 1.02, x = 0.16) # Subtitle plot.  x,y = (0,0) = bottom left, x,y= (1,1) = top right 

#ax.annotate('Asia Pacific Travel Market Jumps', xy=(0.5,0.5), arrowprops=dict(facecolor='red'))
ax.annotate('Asia Pacific Travel Market Jumps 48%', xy=(2, 0.31), xytext=(3, 0.21),arrowprops=dict(facecolor='red'))

for rect in rects1:
    height = rect.get_height()
    ax.annotate('{}'.format(height),
                xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                color = 'lightblue',
                textcoords="offset points",
                ha='center', va='bottom')

for rect in rects2:
    height = rect.get_height()
    ax.annotate('{}'.format(height),
                xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                color = 'blue',
                textcoords="offset points",
                ha='center', va='bottom')
    
plt.tight_layout()
plt.show()
```
