```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from pandas import ExcelWriter
from pandas import ExcelFile
import xlsxwriter
```

```python
h_data = pd.read_excel('Week_6_Homework_Data.xlsx')
```

```python
h_data.head()
```

```python
new_header = h_data.iloc[0] #grab the first row for the header
h_data = h_data[1:] #grab the data from the second row to the end
h_data.columns = new_header #set the header row as the df header
h_data = h_data.dropna(axis='columns') #drop the NaNs
h_data #check the data
```

```python
#setting tick from bottom to the top
plt.rcParams['xtick.bottom'] = plt.rcParams['xtick.labelbottom'] = False 
plt.rcParams['xtick.top'] = plt.rcParams['xtick.labeltop'] = True

plt.rcParams['figure.figsize'] = [8, 10]
fig, ax = plt.subplots()

phil_g = ax.barh(h_data['City'], h_data['Number of Square Feet per $1,000,000'], color = 'lightgrey') #plotting the graph 
phil_g[28].set_color('blue') #changing the Philadelphia bar to blue color 

#for loop through the list of cities from 30 to 39 to change the bar colors grey
for i in phil_g[29:]:
    i.set_color('grey')
        
# removing spines from bottom, left and right         
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
ax.tick_params( axis='x',direction='in') # setting direction og ticks on the top

plt.title('Philadelphia is in top third of home value size per $100,000 spent' ,
          fontsize = 18, ha='center', y = 1.05, x=0.502) # Title plot.  x,y = (0,0) = bottom left, x,y= (1,1) = top right 
plt.show()
```
